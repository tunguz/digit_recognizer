```python
import numpy as np
import pandas as pd
import gc
import time

from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split

from tqdm import tqdm

import warnings
from time import time
```

```python
%%time
train = pd.read_csv('../input/train.csv.gz')
test = pd.read_csv('../input/test.csv.gz')
```

```python
train.head()
```

```python
test.head()
```

```python
features = train.columns[1:]
```

```python
features
```

```python
target = train[['label']].values
train.drop(['label'], axis=1, inplace=True)
```

```python
%%time
train_test = pd.concat([train, test], axis =0)
RS = RobustScaler()
RS.fit(train_test)
del train_test
gc.collect()
gc.collect()
train[features] = RS.transform(train)
test[features] = RS.transform(test)
del RS
gc.collect()
gc.collect()
```

```python
train.head()
```

```python
xgtrain, xgval, y_train, y_val = train_test_split(train, target, test_size=0.2, random_state=42)
```

```python
xgtrain['target'] = y_train
xgval['target'] = y_val
```

```python
xgtrain.to_csv('../input/xgtrain.csv', index=False)
xgval.to_csv('../input/xgval.csv', index=False)
```
